```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
A = ['Kopiko','Biskuit','Tango']
B = [20,40,40]
AB = {'product':A,
        'barang':B}

print(AB)
```

```python
dt = pd.DataFrame(data=AB, columns=['product','barang']);

print(dt)
```

```python
#plt = pd.DataFrame.plot()
dt.plot(x='product', y='barang')
#plt.show()
```

```python


dt.groupby(['product']).sum().plot(kind='pie', y='barang', autopct='%1.0f%%')


```
